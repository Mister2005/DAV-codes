```python
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
```

```python
df['Sal'].fillna(df['Sal'].mean(), inplace=True)
```

```python
df['Emp_ID'] = df['Emp_ID'].fillna(df['Emp_ID'].mode()[0])
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Dept'] = df['Dept'].fillna(df['Dept'].mode()[0])
```

```python
df['Experience'] = df['Experience'].fillna(df['Experience'].median())
```

```python
df['Name'] = df['Name'].fillna(df['Name'].mode()[0])
```

```python
df['Per_score'] = df['Per_score'].fillna(df['Per_score'].median())
```

```python
df['Joining_Date'] = df['Joining_Date'].fillna(df['Joining_Date'].mode()[0])
```
